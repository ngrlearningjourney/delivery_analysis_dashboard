Adding one column named order_difference, which will contain the difference between actual and estimation delivery time  

```python
df['order_difference'] = df['delivery_time_result_in_hours'] - df['delivery_time_estimation_in_hours']
```

#EDA

Preparing for exploratory data analysis, classifying every delivery to their class (on time, delay, early)

```python
df_complete = df[df['order_difference'] == 0]
df_complete["order_result"] = "on time"
```

```python
df_delay = df[df['order_difference'] > 0]
df_delay["order_result"] = "delay"
```

```python
df_early = df[df['order_difference'] < 0]
df_early["order_result"] = "early"
```

```python
df=pd.concat([df_complete, df_delay, df_early], axis=0)
```

```python
df.info()
```

```python
df.groupby(['delivery_status']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, there are 496 complete delivery, 494 damage delivery, 10 lost delivery

```python
df.groupby(['delivery_status','delivery_type']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, all delivery types have a similar number of damaged deliveries (around 119–128) and lost deliveries (ranging from 1–4 deliveries).

```python
df.groupby(['delivery_status','delivery_way']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, **Rail** has the smallest number of completed and damaged deliveries compared to the other delivery methods. For damaged deliveries, **Air, Mixed Way, Road, and Sea** have a similar number of deliveries. For lost deliveries, **Mixed Way** has the highest number.

```python
df.groupby(['delivery_status','product_type']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, all product types have a similar number of damaged and lost deliveries

```python
df.groupby(['product_type']).agg(
    number_of_delivery = ('delivery_id','count')
).sort_values(by="number_of_delivery",ascending=False)
```

According to the data, Groceries is the most commonly delivered product type, while Clothing is the least commonly delivered product type.

```python
df_complete = df[df['delivery_status'] == 'Complete']
df_complete[df_complete['order_result'] == "delay"].groupby(['order_result']).agg(
    delivery_delay_on_average = ('order_difference','median')
)
```

According to the data, the average delivery delay is around **8 hours**.

```python
df_complete[df_complete['order_result'] == "early"].groupby(['order_result']).agg(
    delivery_delay_on_average = ('order_difference','median')
)
```

According to the data, the average early delivery is around **3 hours**.

```python
df_complete.groupby(['order_result']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

according to the data there are 361 number delay deliveries, 110 early deliveries, 25 on time deliveries

```python
df_complete.groupby(['order_result','product_type']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, all product type have similiar number of delay delivery

```python
df_complete.groupby(['order_result','delivery_way']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, **Rail** has the smallest number of delayed deliveries compared to **Air, Mixed Way, Road, and Sea**. For early deliveries, **Air, Mixed Way, Road, and Sea** have a similar number of deliveries. For on-time deliveries, **Mixed Way** has the highest number compared to **Air, Road, and Sea**.

```python
df_complete.groupby(['order_result','delivery_type']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, Economy, Same Day, and Standard delivery types have the highest number of delayed deliveries (90–99 delayed deliveries). Express and Same Day delivery types have the highest number of on-time deliveries. For early deliveries, the Economy delivery type has the highest number of early deliveries.

```python
df_complete.groupby(['order_result','place_end']).agg(
    number_of_delivery = ('delivery_id','count')
)
```

According to the data, all destination locations have a similar number of delayed and on-time deliveries.

```python
df_complete.groupby(['delivery_type']).agg(
    rating_on_average = ('rating','mean')
)
```

According to the data, all delivery type have a similar average of rating

```python
df_complete.groupby(['delivery_type','product_type']).agg(
    rating_on_average = ('rating','mean')
)
```

According to the data, sending **medicine** using **Economy** delivery has the lowest rating compared to the other product types. For **Express** and **Same Day** delivery, sending **machinery** has the lowest rating compared to the other product types. For **Standard** delivery, sending **electronics, documents, groceries, and medicine** has the lowest rating.
